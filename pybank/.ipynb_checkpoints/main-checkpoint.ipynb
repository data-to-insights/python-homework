```python
# PyBank - Pani Maddi

import csv

# Initialize variables
rows = 0
profit = 0
change = 0

# List of stuff
budget = []
newlist = []

# Average function
def Average(lst): 
        return sum(lst) / len(lst) 


# Read the data - "budget_data.csv"
# Total number of months


with open('budget_data.csv') as csvfile:
    readCSV = csv.reader(csvfile)        
    for row in readCSV:
        rows = rows + 1
months = rows -1    
print("     FINANCIAL ANALYSIS      ")
print("_____________________________")
print(f"Number of Months = {months}")    


# Total amount of Profit/Losses over the entire period
with open('budget_data.csv') as csvfile:    
    readCSV = csv.reader(csvfile)    
    is_1st_row = True
    for row in readCSV:
        if is_1st_row:
            is_1st_row = False
            continue
        profit = profit + int(row[1])
    
    print(f"Total Profit/Losses = ${profit}")        
    
    
# Average of the changes in Profit/Losses over the entire period

with open('budget_data.csv') as csvfile:    
    readCSV = csv.reader(csvfile)    
    is_1st_2rows = True
    for row in readCSV:
        if is_1st_2rows:
            is_1st_2rows = False
            continue
        budget.append(row)
            
    for index in range(len(budget)-1):
        change = int(budget[index + 1][1]) - int(budget[index][1])
        newlist.append(change)
                
print(f"Average of the change in Profit/Loss = {round(Average(newlist),2)}")

 
# Greatest increase in profits (date and amount) over the entire period

Maximum_Profit = max(newlist)

for index in range(len(newlist)-1):
    if newlist[index] == Maximum_Profit: 
        Max_row = index

print(f"Greatest increase in profits: {budget[Max_row + 1][0]},  $({Maximum_Profit})")

# Greatest decrease in losses (date and amount) over the entire period

Maximum_Loss = min(newlist)

for index in range(len(newlist)-1):
    if newlist[index] == Maximum_Loss: 
        Min_row = index
        
print(f"Greatest decrease in profits: {budget[Min_row + 1][0]}, $({Maximum_Loss})")

```
