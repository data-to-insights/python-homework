## Imports and Data Preparation

In this section, I prepared the training and testing data for the model. The model will use a rolling 10 day window to predict the 11th day closing price.

1. Used the `window_data` function to generate the X and y values for the model.
2. Split the data into 70% training and 30% testing
3. Apply the MinMaxScaler to the X and y values
4. Reshape the X_train and X_test data for the model using the required input format for the LSTM:

```python
reshape((X_train.shape[0], X_train.shape[1], 1))
```
5. Fear & Greed Model parameters: X1, y1, model1
6. Closing Price Moving Average Model parameters: X2, y2, model2

````python
import numpy as np
import pandas as pd
import hvplot.pandas
%matplotlib inline
````

````python
# Use the MinMaxScaler to scale data between 0 and 1.
from sklearn.preprocessing import MinMaxScaler
````

````python
# Set the random seed for reproducibility
# Note: This is for the homework solution, but it is good practice to comment this out and run multiple experiments to evaluate your model
from numpy.random import seed
seed(1)
from tensorflow import random
random.set_seed(2)
````

````python
# Imports from tensorflow
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
````

````python
# MinMaxScaler function to scale the data between 0 and 1. 
scaler = MinMaxScaler()
````

````python
# Load the fear and greed sentiment data for Bitcoin
df = pd.read_csv('btc_sentiment.csv', index_col="date", infer_datetime_format=True, parse_dates=True)
df = df.drop(columns="fng_classification")

# Load the historical closing prices for Bitcoin
df2 = pd.read_csv('btc_historic.csv', index_col="Date", infer_datetime_format=True, parse_dates=True)['Close']
df2 = df2.sort_index()

# Join the data into a single DataFrame
df = df.join(df2, how="inner")
df.tail()
````

### FEAR & GREED MODEL

````python
# This function accepts the column number for the features (X) and the target (y)
# It chunks the data up with a rolling window of Xt-n to predict Xt
# It returns a numpy array of X any y
def window_data(df, window, feature_col_number, target_col_number):
    X1 = []
    y1 = []
    for i in range(len(df) - window - 1):
        features = df.iloc[i:(i + window), feature_col_number]
        target = df.iloc[(i + window), target_col_number]
        X1.append(features)
        y1.append(target)
    return np.array(X1), np.array(y1).reshape(-1, 1)
````

````python
# Predict Closing Prices using a 10 day window of fear and greed index values and a target of the 11th day closing price
# Used window size = 1
window_size = 1

# Column index 0 is the `fng_value` column; Column index 1 is the 'Close' column
feature_column = 0
target_column = 1
X1, y1 = window_data(df, window_size, feature_column, target_column)
````

````python
# Use 70% of the data for training and the remainder for testing
split = int(0.7 * len(X1))
````

````python
# Split between train and test

X1_train = X1[: split - 1]
X1_test = X1[split:]
y1_train = y1[: split - 1]
y1_test = y1[split:]
````

````python
# Use MinMaxScaler to scale the data between 0 and 1. 

scaler.fit(X1)
X1_train = scaler.transform(X1_train)
X1_test = scaler.transform(X1_test)

scaler.fit(y1)
y1_train = scaler.transform(y1_train)
y1_test = scaler.transform(y1_test)
````

````python
# Reshape the features for the model

X1_train = X1_train.reshape((X1_train.shape[0], X1_train.shape[1], 1))
X1_test = X1_test.reshape((X1_test.shape[0], X1_test.shape[1], 1))
````

---

## Build and Train the LSTM RNN for "FEAR & GREED" MODEL

````python
# Build the LSTM model. 
# The return sequences need to be set to True if you are adding additional LSTM layers, but 
# You don't have to do this for the final layer. 
# Define the LSTM RNN model.
model1 = Sequential()

number_units = 5
dropout_fraction = 0.2

# Layer 1
model1.add(LSTM(units=number_units,return_sequences=True,input_shape=(X1_train.shape[1], 1)))
model1.add(Dropout(dropout_fraction))
# Layer 2
model1.add(LSTM(units=number_units, return_sequences=True))
model1.add(Dropout(dropout_fraction))
# Layer 3
model1.add(LSTM(units=number_units))
model1.add(Dropout(dropout_fraction))
# Output layer
model1.add(Dense(1))
````

````python
# Compile the model
model1.compile(optimizer="adam", loss="mean_squared_error")
````

````python
# Summarize the model
model1.summary()
````

````python
# Train the FEAR & GREED model

model1.fit(X1_train, y1_train, epochs=10, shuffle=False, batch_size=1, verbose=1)
````

---

## Fear & Greed - Model Performance

````python
# Evaluate the model
model1.evaluate(X1_test, y1_test)
````

````python
# Make some predictions
predicted1 = model1.predict(X1_test)
````

````python
# Recover the original prices instead of the scaled version
predicted_fng = scaler.inverse_transform(predicted1)
real_prices = scaler.inverse_transform(y1_test.reshape(-1, 1))
````

### CLOSING PRICE (10day rolling) MODEL

````python
# This function accepts the column number for the features (X) and the target (y)
# It chunks the data up with a rolling window of Xt-n to predict Xt
# It returns a numpy array of X any y
def window_data(df, window, feature_col_number, target_col_number):
    X2 = []
    y2 = []
    for i in range(len(df) - window - 1):
        features = df.iloc[i:(i + window), feature_col_number]
        target = df.iloc[(i + window), target_col_number]
        X2.append(features)
        y2.append(target)
    return np.array(X2), np.array(y2).reshape(-1, 1)
````

````python
# Predict Closing Prices using a 10 day window of fear and greed index values and a target of the 11th day closing price
# Used window size = 1
window_size = 1

# Column index 0 is the `fng_value` column; Column index 1 is the 'Close' column
feature_column = 1
target_column = 1
X2, y2 = window_data(df, window_size, feature_column, target_column)
````

````python
# Split between train and test

X2_train = X2[: split - 1]
X2_test = X2[split:]
y2_train = y2[: split - 1]
y2_test = y2[split:]
````

````python
# Use MinMaxScaler to scale the data between 0 and 1. 

scaler.fit(X2)
X2_train = scaler.transform(X2_train)
X2_test = scaler.transform(X2_test)

scaler.fit(y2)
y2_train = scaler.transform(y2_train)
y2_test = scaler.transform(y2_test)
````

````python
# Reshape the features for the model

X2_train = X2_train.reshape((X2_train.shape[0], X2_train.shape[1], 1))
X2_test = X2_test.reshape((X2_test.shape[0], X2_test.shape[1], 1))
````

## Build and Train the LSTM RNN for "CLOSING PRICE 10day AVG" MODEL

````python
# Build the LSTM model. 
# The return sequences need to be set to True if you are adding additional LSTM layers, but 
# You don't have to do this for the final layer. 
# Define the LSTM RNN model.
model2 = Sequential()

number_units = 5
dropout_fraction = 0.2

# Layer 1
model2.add(LSTM(units=number_units,return_sequences=True,input_shape=(X2_train.shape[1], 1)))
model2.add(Dropout(dropout_fraction))
# Layer 2
model2.add(LSTM(units=number_units, return_sequences=True))
model2.add(Dropout(dropout_fraction))
# Layer 3
model2.add(LSTM(units=number_units))
model2.add(Dropout(dropout_fraction))
# Output layer
model2.add(Dense(1))
````

````python
# Compile the model
model2.compile(optimizer="adam", loss="mean_squared_error")
````

````python
# Summarize the model
model2.summary()
````

````python
# Train the CLOSING price 10day Moving AVG model

model2.fit(X2_train, y2_train, epochs=10, shuffle=False, batch_size=1, verbose=1)
````

## CLOSING price 10day Moving AVG - Model Performance

````python
# Evaluate the model
model2.evaluate(X2_test, y2_test)
````

````python
# Make some predictions
predicted2 = model2.predict(X2_test)
````

````python
# Recover the original prices instead of the scaled version
predicted_closing = scaler.inverse_transform(predicted2)
````

````python
### DataFrame with Actual values, Fear & Greed prediction, Closing Price MA prediction
### Fear & Greed predictions are very far from Actuals
### Closing Price MA predictions are much closer to Actuals
````
